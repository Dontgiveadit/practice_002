```python
import pandas as pd

# สร้างข้อมูลตัวอย่าง (สมมุติเป็นข้อมูลผู้ป่วยง่ายๆ)
data = {
    'Patient_ID': [1, 2, 3, 4, 5],
    'Age': [55, 62, 48, 70, 59],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
    'Cholesterol_Level': [220, 190, 250, 210, 235]
}

# สร้าง DataFrame
df = pd.DataFrame(data)

# แสดง DataFrame
print(df)
```

```python
# หาค่าเฉลี่ยของระดับคอเลสเตอรอล
average_cholesterol = df['Cholesterol_Level'].mean()
print(f"\nAverage Cholesterol Level: {average_cholesterol}")

# นับจำนวนผู้ป่วยชายและหญิง
gender_counts = df['Gender'].value_counts()
print(f"\nGender Distribution:\n{gender_counts}")
```

```python
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem.Draw import IPythonConsole

# สร้างโมเลกุลจาก SMILES string (ตัวอย่าง: Aspirin)
mol = Chem.MolFromSmiles('CC(=O)Oc1ccccc1C(=O)O')

# แสดงโครงสร้างโมเลกุล (จะแสดงภาพใน Colab)
mol

# พิมพ์ SMILES กลับมา
print(Chem.MolToSmiles(mol))

# คำนวณจำนวนอะตอม
print(f"Number of atoms: {mol.GetNumAtoms()}")
```

```python
import plotly.express as px
import pandas as pd

# สร้าง DataFrame ตัวอย่าง
df = pd.DataFrame({
    "Year": [2000, 2001, 2002, 2003, 2004],
    "Drug_A_Sales": [100, 120, 150, 130, 180],
    "Drug_B_Sales": [80, 90, 110, 100, 140]
})

# สร้างกราฟเส้นแบบ Interactive
fig = px.line(df, x="Year", y=["Drug_A_Sales", "Drug_B_Sales"], title="Drug Sales Over Years")
fig.show()
```

```python
from transformers import pipeline

# โหลดโมเดลสำหรับทำ Sentiment Analysis (การวิเคราะห์ความรู้สึก)
classifier = pipeline("sentiment-analysis")

# ลองวิเคราะห์ข้อความ
result = classifier("This drug shows promising results in clinical trials.")
print(result)

result2 = classifier("The side effects of this treatment are quite severe.")
print(result2)
```
